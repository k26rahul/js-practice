```python
import pandas as pd

df = pd.read_csv('ca212fb975d63bfff9687874de60bd2f67cc054d6e26c-job-profiles.csv')

len(df)
```

```python
df_female = df[df['Gender'] == 'F']
df = df_female

len(df)
```

```python
df[~df['Email'].str.endswith('@ds.study.iitm.ac.in')][['Name', 'Email', 'Mobile', 'Score']]
```

```python
filtered_df = df
# filtered_df = df[df['Email'].str.endswith('@ds.study.iitm.ac.in')]


len(filtered_df)
```

```python
sorted_df = filtered_df.sort_values(by='Score', ascending=False)
sorted_df = sorted_df[['Name', 'Email', 'Mobile', 'Score']]
len(sorted_df)
sorted_df
```

```python
# Convert Mobile to string and remove '91' prefix if present
sorted_df['Mobile'] = sorted_df['Mobile'].astype(str).apply(lambda x: x[2:] if x.startswith('91') else x)

sorted_df
```

```python
g_form_df = pd.read_csv('stree_sakti_quiz_participants.csv')

g_form_emails = list(g_form_df['Email Address'])
g_form_emails[:5]

g_form_whatsapp_numbers = g_form_df['Whatsapp Number'].tolist()
g_form_whatsapp_numbers[:5]  # Display the first 5 numbers

g_form_whatsapp_numbers = [str(x) for x in g_form_whatsapp_numbers]
```

```python
not_in_g_form_df = sorted_df[~sorted_df['Email'].isin(g_form_emails)]

len(not_in_g_form_df)
not_in_g_form_df

not_in_g_form_df.to_csv('not_in_g_form_df.csv')
```

```python
g_form_df = sorted_df[sorted_df['Email'].isin(
    g_form_emails) | sorted_df['Mobile'].astype(str).isin(g_form_whatsapp_numbers)]

len(g_form_df)
g_form_df

g_form_df.to_csv('all.csv')
```

```python
top_10 = g_form_df.nlargest(10, 'Score')
top_10
```

```python
top_10.to_csv('top_10.csv', index=False)

top_10_emails = top_10[['Email']]
top_10_emails.to_csv('top_10_emails.csv', index=False)
```
