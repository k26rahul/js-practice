```python
import pandas as pd

df = pd.read_csv('ca212fb975d63bfff9687874de60bd2f67cbf06c44558-job-profiles.csv')

df.columns
```

```python
df_female = df[df['Gender'] == 'F']
len(df_female)

df = df_female
```

```python
sorted_df = df.sort_values(by='Score', ascending=False)
sorted_df = sorted_df[['Name', 'Email', 'Mobile', 'Score']]
len(sorted_df)
sorted_df
```

```python
filtered_df = sorted_df[sorted_df['Email'].str.endswith('@ds.study.iitm.ac.in')]
len(filtered_df)
filtered_df
```

```python
g_form_emails = pd.read_csv('stree_sakti_quiz_participants.csv')
g_form_emails = list(g_form_emails['Email Address'])
g_form_emails[:5]
```

```python
not_in_g_form_df = filtered_df[~filtered_df['Email'].isin(g_form_emails)]
not_in_g_form_df
```

```python
not_in_g_form_df.to_csv('not_in_g_form_df.csv')
```

```python
filtered_g_form_df = filtered_df[filtered_df['Email'].isin(g_form_emails)]
len(filtered_g_form_df)
filtered_g_form_df
```

```python
top_10 = filtered_g_form_df.nlargest(10, 'Score')
top_10

top_10.to_csv('top_10.csv', index=False)
```

```python
top_10_emails = top_10[['Email']]
top_10_emails.to_csv('top_10_emails.csv', index=False)
```
